```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Өгөгдөл бэлдэх 100 ширхэг x, y үүсгэх
# y = w*X + b
X = np.random.normal(0, 1, 100)*10
y = 2*X + np.random.normal(0, 3, 100) +10
plt.plot(X,y,"o")
```

```python
#Загвар бэлдэх
# y = w*x + b
class LinearRegress():
  def __init__(self, learning_rate):
    self.w = np.random.randn(1) #1 indicates the dimension of parameter
    self.b = np.random.randn(1) 
    self.learning_rate = learning_rate#0.01
  
  def forward(self, x):
    # y = w*x+b
    y_pred = self.w*x +self.b
    return y_pred

  def get_params(self):
    print(self.w)
    print(self.b)

  def get_loss(self, x, y):
    y_pred = self.forward(x)
    S = np.sum ((y_pred - y)**2)/len(x)
    return S

  def get_grad(self, x, y):
    y_pred = self.forward(x)
    grad_w = np.sum(x.dot((y_pred - y)))/len(x)
    grad_b = np.sum(y_pred - y)/len(x)
    return grad_w, grad_b

# Optimization algorith: gradient descent that updates w & b such that 
# w<- w: learning_rate * grad_w
# w<- b: learning_rate * grad_b
  def backward(self, grad_w, grad_b):
    self.w = self.w - self.learning_rate*grad_w
    self.b = self.b - self.learning_rate*grad_b

```

```python
#Загварын instance үүсгэх
model = LinearRegress(0.01)
y_pred = model.forward(X)
model.get_params()
print(X[:10])
print(y_pred[:10])
```

```python
#Загвар сургах
model.get_params()
epoch = 1000
losses = list()
for i in range(epoch):
  grad_w, grad_b = model.get_grad(X, y)
  model.backward(grad_w, grad_b)
  loss = model.get_loss(X, y)
  losses.append(loss)

  if i % 100==0:
    print(i, loss)
    plt.scatter(X, y)
    x_1 = np.array([-30,30])
    y_1 = model.w*x_1 + model.b
    plt.plot(x_1,y_1,'r')
    plt.show()

model.get_params()
plt.plot(range(epoch), losses)
plt.xlabel("Epoch")
plt.ylabel("Losses")
```
